```python
#importing the required libraries

import pandas as pd
import numpy as np
```

```python
#Since there are so many sheets, we will write a function which will return the dataframe 

def fileprocessing(filename,sheetname):
    df = pd.read_excel(filename, sheet_name = sheetname) #read the file in and specify the columns in the next step
    df.columns = ['Year','Year_Mon','Division_Descriptions','Country','Exports in Euros','Exports in Tonnes','Imports in Euros','Imports in Tonnes','Total Euros','Total Tonnes']
    df = df.drop(labels=range(0,4), axis=0) #the file contains certain rows as empty, so drop them.
    df['Year'] = df['Year'].fillna(method='ffill') #forward fill the year, year_month and division descriptions
    df['Year_Mon'] = df['Year_Mon'].fillna(method='ffill')
    df['Division_Descriptions'] = df['Division_Descriptions'].fillna(method='ffill')
    df['Year'] = df['Year'].str.strip() #strip the space in the year column 
    df['Year'] = df['Year'].str.strip('Jan-Dec') #strip the string other than year number
    df["Year_Mon"] = df["Year_Mon"]+" "+df["Year"] #replace the year_mon column with month and year
    df["Division_id"] = df["Division_Descriptions"].str[:2] #create a division id
    df["Division_Descriptions"] = df["Division_Descriptions"].str[3:] #replace the column with the fields but not numbers
    cols1 = ['Exports in Euros','Exports in Tonnes','Imports in Euros','Imports in Tonnes','Total Euros','Total Tonnes']
    df.loc[:,cols1] = df.loc[:,cols1].replace(np.NaN,0) #replace the NaN values with zero in the above listed columns
    return df   #return the processed dataframe
```

```python
#In this excel file, it contains multiple sheets which we can list them and assign it to a variable sheet_name

sheet_name = ['2010-2012','2013-2015','2016-2019','2020']
df_10_20 = []     #create an empty list in which we will append the dataframes 
for sheets in range(len(sheet_name)): #create the for loop, process all the sheets and append them in the empty list
    df = fileprocessing('Manoj01_04_1.xlsx',sheets)  #bring the fileprocessing function to process all the sheets
    df_10_20.append(df)    #append them to a empty list
```

```python
#concat the appended df list into a single dataframe

df = pd.concat(df_10_20)
df
```
